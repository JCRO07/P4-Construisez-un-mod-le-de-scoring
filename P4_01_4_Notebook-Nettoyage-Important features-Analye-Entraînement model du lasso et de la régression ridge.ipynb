# Modèles du lasso et de la régression ridge

```python
# one-hot encoding of categorical variables
xtrain = pd.get_dummies(xtrain)
xtest = pd.get_dummies(xtest)

print('Training Features shape: ', xtrain.shape)
print('Testing Features shape: ', xtest.shape)
```

```python
xtrain
```

```python
xtrain1 = xtrain
xtrain1
```

```python
xtrain1 = xtrain1.drop(xtrain1.columns[[1]], axis='columns')
xtrain1 
```

```python
#X_train = xtrain1.iloc[:2500,0:247]
#y_train = xtrain.iloc[:2500,1]
#X_test = xtrain1.iloc[2500:,0:247]
#y_test = xtrain.iloc[2500:,1]
```

```python
X = xtrain1.values
y = xtrain['TARGET'].values
```

```python
#Séparation des données en un jeu d’entraînement et un jeu de test. Le jeu de test sera stratifié


from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, stratify=y, test_size=0.7, random_state=42)
```

```python
#Standardardisation des données d’entraînement et application de la même transformation aux données de test 
from sklearn import preprocessing
std_scale = preprocessing.StandardScaler().fit(X_train)
X_train_std = std_scale.transform(X_train)
X_test_std = std_scale.transform(X_test)
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
X_train
```

```python
from sklearn import linear_model
import numpy as np

# On crée un modèle de régression linéaire
lr = linear_model.LinearRegression()

# On entraîne ce modèle sur les données d'entrainement
lr.fit(X_train,y_train)

# On récupère l'erreur de norme 2 sur le jeu de données test comme baseline
baseline_error = np.mean((lr.predict(X_test) - y_test) ** 2)

print(baseline_error)
```

```python
y_test_pred1 = lr.predict(X_test)
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import PrecisionRecallDisplay

prec, recall, _ = precision_recall_curve(y_test, y_test_pred1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
#y_score = clf.decision_function(X_test)

fpr, tpr, _ = roc_curve(y_test, y_test_pred1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```

```python
n_alphas = 200
alphas = np.logspace(-5, 5, n_alphas)
```

```python
from sklearn.linear_model import Ridge
ridge = linear_model.Ridge()

coefs = []
errors = []
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(X_train, y_train)
    coefs.append(ridge.coef_)
    errors.append([baseline_error, np.mean((ridge.predict(X_test) - y_test) ** 2)])
```

```python
y_test_pred2 = ridge.predict(X_test)
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import PrecisionRecallDisplay

prec, recall, _ = precision_recall_curve(y_test, y_test_pred2)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
#y_score = clf.decision_function(X_test)

fpr, tpr, _ = roc_curve(y_test, y_test_pred2)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```

```python
import matplotlib.pyplot as plt

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

```python
ax = plt.gca()

ax.plot(alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.axis('tight')
plt.show()
```

```python
min(errors)
```

```python
n_alphas = 300
alphas = np.logspace(-5, 1, n_alphas)
lasso = linear_model.Lasso(fit_intercept=False)

coefs = []
errors = []
for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(X_train, y_train)
    coefs.append(lasso.coef_)
    errors.append([baseline_error, np.mean((lasso.predict(X_test) - y_test) ** 2)])
```

```python
y_test_pred3 = lasso.predict(X_test)
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import PrecisionRecallDisplay

prec, recall, _ = precision_recall_curve(y_test, y_test_pred3)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
#y_score = clf.decision_function(X_test)

fpr, tpr, _ = roc_curve(y_test, y_test_pred3)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```

```python
ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.axis('tight')
plt.show()
```

```python
ax = plt.gca()

ax.plot(alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.axis('tight')
plt.show()
```
