#  Régression logistique

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn import metrics

lr = LogisticRegression(solver = 'liblinear')

lr.fit(X_train,y_train)

# On récupère la prédiction de la valeur positive
y_prob = lr.predict_proba(X_test)[:,1] 

# On créé un vecteur de prédiction à partir du vecteur de probabilités
y_pred = np.where(y_prob > 0.5, 1, 0) 

false_positive_rate1, true_positive_rate1, thresholds1 = roc_curve(y_test, y_prob)
roc_auc1 = auc(false_positive_rate1, true_positive_rate1)
print(roc_auc1)
print("AUC1 : {:.2f}".format(roc_auc1))
print("RMSE1 : {:.2f}".format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import PrecisionRecallDisplay
print("AUC1 : {:.2f}".format(roc_auc1))
print("RMSE1 : {:.2f}".format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
prec, recall, _ = precision_recall_curve(y_test, y_prob,
                                         pos_label=lr.classes_[1])
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
#y_score = clf.decision_function(X_test)
print("AUC1 : {:.2f}".format(roc_auc1))
print("RMSE1 : {:.2f}".format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
fpr, tpr, _ = roc_curve(y_test, y_prob, pos_label=lr.classes_[1])
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```
